# 로지스틱(시그모이드) 함수

```python
data = [[2,0],[4,0],[6,0],[8,1],[10,1],[12,1],[14,1]]

x_data_ = np.array(data)[:,0]
y_data_ = np.array(data)[:,1]
print(x_data_)
print(y_data_)
plt.scatter(x_data_, y_data_)
plt.show()
```

```python
# 기울기와 절편 찾기
# 기울기 a와 절편 b 초기화
a = 0
b = 0

# 학습률
lr = 0.05

# 시그모이드 함수 정의
def sigmoid(x):
    return 1 / (1 + np.e**(-x))

# 경사 하강법 실행
for i in range(2001):
    for x_data, y_data in data:
        # a에 관한 편미분
        a_diff = x_data*(sigmoid(a*x_data+b)-y_data)
        
        # b에 관한 편미분
        b_diff = sigmoid(a*x_data+b) - y_data
        a = a - lr * a_diff
        b = b - lr * b_diff
        
        # 에포크(epoch)가 1000번일 때만 출력
        if i % 1000 == 0:
            print("epoch=%.f, 기울기=%.04f, 절편=%.04f" % (i,a,b))
            
```

```python
# 경사하강법으로 구한 기울기와 절편으로 그래프 그리기
plt.scatter(x_data_, y_data_)
plt.xlim(0, 15)
plt.ylim(-.1, 1.1)
x_range = np.arange(0, 15, 0.1)
plt.plot(x_range, sigmoid(a*x_range+b), c='r')
plt.show()
```

# 다층 퍼셉트론

```python
# 가중치 설정
w11 = np.array([-2,-2])
w12 = np.array([2,2])
w2 = np.array([1,1])

# 바이어스 설정
b1 = 3
b2 = -1
b3 = -1
```

```python
def MLP(x, w, b):
    y = np.sum(x*w) + b
    
    if y > 0:
        return 1
    else:
        return 0     
```

```python
## n2
def OR(x1, x2):
    return MLP(np.array([x1, x2]), w12, b2)

## n1
def NAND(x1, x2):
    return MLP(np.array([x1, x2]), w11, b1)

## y
def AND(x1, x2):
    return MLP(np.array([x1, x2]), w2, b3)

##
def XOR(x1, x2):
    return AND(NAND(x1, x2), OR(x1, x2))
```

```python
for x in [(0,0),(1,0),(0,1),(1,1)]:
    y = XOR(x[0],x[1])
    print('입력값:',str(x), '출력값:',str(y))
```
